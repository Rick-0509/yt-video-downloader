#### CAMBIAR EL LINK PARA PODER DESCARGAR CUALQUIER TIPO DE VIDEO DE YOUTUBE

```python
import yt_dlp

def descargar_video(url, carpeta_destino="."):
    opciones = {
        'outtmpl': f'{carpeta_destino}/%(title)s.%(ext)s',
        'format': 'best[ext=mp4]/best',  # ✅ Mejor calidad en UN solo archivo, sin necesitar FFmpeg
        'noplaylist': True,
    }

    with yt_dlp.YoutubeDL(opciones) as ydl:
        info = ydl.extract_info(url, download=False)
        print(f"🎬 Título    : {info['title']}")
        print(f"📺 Resolución: {info.get('height', 'N/A')}p")
        print("⬇️  Descargando...\n")
        ydl.download([url])
        print("✅ Descarga completada!")

descargar_video("https://www.youtube.com/watch?v=5JJwESLVTjs&list=RD5JJwESLVTjs&start_radio=1")
```
